```python
import pandas as pd
import numpy as np
from hazm import *
from tf_idf_with_max_padding import word_tokenizing, unique_tokens
```

```python
weather_list = [ 'هوا', 'دما', 'هواشناسی' ,'باران' ,'آفتاب' ,'آسمان' ,'بارون' ,'سرد' ,'گرم' ,'بارش' ,
                'برف' ,'ابری' ,'طوفان' ]
```

```python
shari_list = ['اذان', 'طلوع', ' اوقات شرعی', 'روزه', 'نیمه شب شرعی', 'افق', 'نماز', 'سحر', 'افطار', 'غروب', 'مغرب',
              'قضا', 'اذون', 'نیمه‌شب شرعی', 'رمضون', 'نیمه‌شب']
```

```python
time_list = ['ساعت', 'زمان', 'فاصله‌ی زمانی', 'نیمه شب', 'دقیقه', 'شب', 'روز', 'وقت محلی', 'محلی', 'اختلاف زمان',
             'زمانی', 'اختلاف ساعت']
```

```python
answers = pd.read_csv('answers_clean.csv')
classes = answers['class']
```

```python
def predict (tokens, lemmatizer):
    pred_classes = []
    for token in tokens or lemmatizer.lemmatize(token) in tokens:
        if token in weather_list : 
            pred_classes.append(1)
        elif token in shari_list : 
            pred_classes.append(2)
        elif token in time_list : 
            pred_classes.append(3)
#         else :
#             pred_classes.append(4)
    return unique_tokens(pred_classes)
```

```python
def predict_for_all ():
    y_pred = []
    tokens, _ = word_tokenizing()
    lemmatizer = Lemmatizer
    for token in tokens :
        y_pred.append(predict(token, lemmatizer))
    return y_pred
```

```python
y_pred = predict_for_all()
```
